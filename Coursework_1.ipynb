**1. Мердж всех датасетов кроме students.csv**

```python
#-*- coding: utf-8 -*-
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
courses=pd.read_csv('courses.csv',delimiter=',',encoding='utf-8')
course_contents=pd.read_csv('course_contents.csv',delimiter=',',encoding='utf-8')
progresses=pd.read_csv('progresses.csv',delimiter=',',encoding='utf-8')
progress_phases=pd.read_csv('progress_phases.csv',delimiter=',',encoding='utf-8')
students=pd.read_csv('students.csv',delimiter=',',encoding='utf-8')
del(students['id_'])
df_1=course_contents.merge(courses,how='inner',right_on='id',left_on='course_id')
del (df_1['Unnamed: 0'])
del (df_1['id'])
df_2=progresses.merge(progress_phases,how='inner',left_on='id',right_on='progress_id')


# Проверка правильности проведённого мерджа
df=df_1.merge(df_2,how='inner',on=['course_id','module_number','lesson_number'])
df[df.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

2. *Общее количество курсов в датасете*


```python
print(len(df.groupby('course_id')))
```

3. *Количество модулей на каждом курсе*

```python
df_count_mod=df.loc[:,['title','module_number']].drop_duplicates()
df_count_mod.groupby('title').agg('count')
```

4. *Количество уроков в каждом модуле на каждом курсе*

```python
df_count_lessons=df.loc[:,['title','module_number','lesson_number']].drop_duplicates()
df_count_lessons_1=df_count_lessons.groupby(['title','module_number']).agg('count')
df_count_lessons_1['lesson_number']
```

6. *Количество учеников на каждом курсе*

```python
df_count_people = df.loc[:,['title','student_id']].drop_duplicates().groupby(['title'])['student_id'].agg('count')
df_count_people
```

*Мердж датасета students*

```python
students=students.dropna(axis='index',how='any',subset=['birthday'])
age=lambda x: int((datetime.datetime.now()-datetime.datetime.strptime(x,'%Y-%m-%d')).days/365)
students['age']=students['birthday'].apply(age)
students=students[(students['age']>10) &(students['age']<90)]
df_age=df.merge(students,how='left',left_on='student_id',right_on='id')
del(df_age['id_x'])
del(df_age['id_y'])
```

**3. Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом курсе. График должен иметь заголовок.Значения должны быть отсортированы.Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда).Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно**


```python
df_field=df.loc[:,['title','field']].drop_duplicates()
df_field=df_field.sort_values('title')
df_field['student_count']=list(df_count_people)
df_field.sort_values('student_count',ascending=False,inplace=True)
colors=[]
for i in df_field['field']:
    if i=='Marketing':
        colors.append('r')
    elif i=='':
        colors.append('b')
    elif i=='Development':
        colors.append('g')
    elif i=='Business':
        colors.append('y')
handles=[Rectangle((0,0),1,1,color=c,ec='k') for c in ['r','b','g','y','k']]
labels=['Marketing','Design','Development','Business','Median']
fig, ax = plt.subplots()
ax.barh(df_field['title'],df_field['student_count'], color=colors,alpha=0.1)
ax.set_title(f'Количество студентов на каждом курсе')
line=plt.axvline(x=df_field['student_count'].apply('median'),ls='--',color='k',label='Median')
fig.set_figwidth(20)
fig.set_figheight(10)
ax.set_xlabel('Студенты')
ax.set_ylabel('Курсы')
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False) 
ax.spines['bottom'].set_visible(False) 
ax.spines['left'].set_visible(False)

ax.legend(handles,labels)
plt.show()
```

**1. Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.**

```python
df_students_grow_up=df[(df['is_homework'] == True)&(df['start_date'] is not None)]
df_students_grow_up=df_students_grow_up.loc[:,['title','student_id']]
df_students_grow_up.drop_duplicates(inplace=True)
indexes=list(df_students_grow_up.index)
df_students_grow_up=df.iloc[indexes,[8,11,14]]
df_students_grow_up
format_date='%Y-%m'
date=lambda x: str(x)[:7]

start=lambda x: datetime.datetime.strptime(x.split()[0][:-3],format_date)
df_students_grow_up['start_date']=df_students_grow_up['start_date'].apply(start)
df_students_grow_up=df_students_grow_up[(df_students_grow_up['start_date']>=datetime.datetime.strptime('2016-03',format_date))&(df_students_grow_up['start_date']<datetime.datetime.strptime('2019-08',format_date))]
df_students_grow_up['start_date']=df_students_grow_up['start_date'].apply(date)
df_students_grow_up=df_students_grow_up.groupby(['title','start_date']).agg('count')
df_students_grow_up.sort_values('title',inplace=True)
df_students_grow_up.columns=['Increase in students']
df_students_grow_up
```

**2. На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса. 15 графиков. Графики должны иметь заголовки, оси должны быть подписаны. Ticks нужно развернуть так, чтобы они были читаемы.**

```python
format_2='%Y-%m-%d'
def get_datetime(x):
    return pd.Series([datetime.datetime.strptime(x[0][:10],format_2),datetime.datetime.strptime(x[1][:10],format_2)])
def parsing_min_date(x):
    return min(list(x))
def parsing_max_date(x):
    return max(list(x))
def spent_time(x):
    return (x[1]-x[0]).days
df_problem_module=df[(df['is_homework'] == True)&(df['status'] =='done')]
df_problem_module[['start_date','finish_date']]=df_problem_module[['start_date','finish_date']].apply(get_datetime,axis=1)
df_problem_module=df_problem_module[['title','module_number','lesson_number','student_id','start_date','finish_date']]
df_problem_module=df_problem_module.groupby(['title','module_number','lesson_number','student_id'])[['start_date','finish_date']].agg({'start_date':parsing_min_date,'finish_date':parsing_max_date})
df_problem_module['spent_time']=df_problem_module[['start_date','finish_date']].apply(spent_time,axis=1)
df_problem_module

df_problem_module_=df_problem_module.groupby(['title','module_number','lesson_number'])['spent_time'].agg(['min','max','mean','median'])
df_problem_module_
```
